```python
# Number of Employees in each Department
count = df.groupby('Department')['EmployeeCount'].sum()
ax = sb.barplot(x = count.index, y = count.values)
actual_count = count.values
ax.bar_label(container = ax.containers[0],labels = actual_count)
plt.title('Number of Employees per Department',fontweight = 'bold');
```

```python
# Which department contributes the most to churned employees
churned_employees = df.Attrition.value_counts()['Yes']
churn_employees = df[df.Attrition == 'Yes'].groupby('Department').size()/churned_employees * 100
churn_employees = churn_employees.reset_index()
churn_employees.rename(columns = {0:'churn_percentage'},inplace = True)
label = np.round(churn_employees.churn_percentage.values,2)
label = [str(lbl) + '%' for lbl in label]
ax = sb.barplot(y='Department',x = 'churn_percentage',data = churn_employees)
ax.bar_label(container = ax.containers[0],labels = label)
plt.xlabel('Percentage of Churn Contribution (%)',fontweight='bold')
plt.ylabel('Department',fontweight='bold')
plt.title('Department Contribution to Employee Churn',fontweight='bold')
plt.show();
```

Note :
- This is different from churn rate because the contribution to churn is churned employees in a department / total churned employees
- While churn rate is churned employees in a department/total employees in that department

```python
# Now lets look at the churn and retention rate per Department 

# Calculate churn rate
churn_rate = df[df.Attrition == 'Yes'].groupby('Department').size() / df.groupby('Department')['EmployeeCount'].sum() * 100
churn_rate = churn_rate.reset_index()
churn_rate.rename(columns = {0:'churn_rate'}, inplace = True)

# Calculate retention rate
churn_rate['retention_rate'] = 100 - churn_rate['churn_rate']

# Melt the dataframe to have a single column for rate and a column for the type of rate for easy plotting
churn_rate_melted = churn_rate.melt(id_vars=['Department'], value_vars=['churn_rate', 'retention_rate'], var_name='Rate Type', value_name='Percentage')

# Plot the bar plot
plt.figure(figsize=(20, 20))
ax = sb.barplot(x='Percentage', y='Department', hue='Rate Type', data=churn_rate_melted)
# Obtaining the bar label values
churn = churn_rate_melted[churn_rate_melted['Rate Type'] == 'churn_rate'].Percentage 
retention = churn_rate_melted[churn_rate_melted['Rate Type'] == 'retention_rate'].Percentage
churn = [f'{p[0]:.2f}%' for p in zip(churn) ]
retention = [f'{p[0]:.2f}%' for p in zip(retention) ]
# Add text labels for churn bars
for i in range(len(churn_rate_melted)):
    if churn_rate_melted['Rate Type'][i] == 'churn_rate':
        ax.bar_label(container=ax.containers[0], labels=churn)

# Add text labels for retention bars
for i in range(len(churn_rate_melted)):
    if churn_rate_melted['Rate Type'][i] == 'retention_rate':
        ax.bar_label(container=ax.containers[1], labels=retention)
plt.xlabel('Percentage', fontweight='bold')
plt.ylabel('Department', fontweight='bold')
plt.legend(title='Rate Type')
plt.title('Churn and Retention rates across all departments',fontweight = 'bold',x = 0.45)
plt.show()
```

#### Insights
- The bias with churn contribution it will not favor the department with more employees so churned rate is a better estimate
- The sales department has the highest churn rate which is something to look out for

## Preprocessing 

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, ShuffleSplit
from sklearn.metrics import *
```

```python
df_copy = df.copy()
```

```python
scaler = MinMaxScaler()

scaled_df = pd.DataFrame(df_copy.select_dtypes(exclude='number'))
# Loop through each numeric column and scale it
for column in df_copy.select_dtypes(include=['float64', 'int64']).columns:
    # Fit and transform the column
    scaled_values = scaler.fit_transform(df_copy[[column]])
    # Convert scaled values to DataFrame
    scaled_column = pd.DataFrame(scaled_values, columns=[column])
    # Drop all numeric columns
    df_copy = df_copy.drop(columns = ([column]), axis=1)
    # Concatenate scaled column to scaled DataFrame
    scaled_df = pd.concat([scaled_df,scaled_column], axis=1)
```

```python
le = LabelEncoder()
for col in scaled_df.select_dtypes(exclude='number').columns:
    scaled_df[col]=le.fit_transform(scaled_df[col])
```

```python
scaled_df.head()
```

```python
# Dropping irrelevant columns 
exclude_cols = ['EmployeeCount','StandardHours','Attrition','BusinessTravel','EducationField','OverTime']
# Features
X = scaled_df.drop(columns = (['EmployeeCount','StandardHours','Attrition','BusinessTravel','EducationField','OverTime']),axis = 1)
# Target
Y = scaled_df['Attrition']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
smote = SMOTE(random_state = 14)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

## Model Building

```python
# Importing necssary packages for the ml models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier, XGBRegressor
```

```python
from sklearn.metrics import f1_score, precision_score

def classification(X_tr, y_tr, X_te, y_te, method):
    method.fit(X_tr, y_tr)
    p_train = method.predict(X_tr)
    p_test = method.predict(X_te)
    
    train_f1 = f1_score(y_tr, p_train, average='micro')
    test_f1 = f1_score(y_te, p_test, average='macro')
    
    train_precision = precision_score(y_tr, p_train, average='macro')
    test_precision = precision_score(y_te, p_test, average='macro')
    
    print(f'{method}')
    print('Train Precision Score:', train_precision)
    print('Test Precision Score:', test_precision)
    print('Train F1 Score:', train_f1)
    print('Test F1 Score:', test_f1)
    print('-'*20)


```

```python
xgbc = XGBClassifier(random_state=14)
logreg = LogisticRegression(solver='liblinear')
svm = SVC(kernel='linear')
randomforest = RandomForestClassifier()
methods = [xgbc,logreg,svm,randomforest]
for method in methods:
    classification(X_train,y_train,X_test,y_test,method = method)
```

```python
smote = SMOTE(random_state = 14)
X_test_smote, y_test_smote = smote.fit_resample(X_test, y_test)
```

```python
print('USING SMOTE')
print('_'*50)
for method in methods:
    classification(X_train_smote,y_train_smote,X_test_smote,y_test_smote,method = method)
```

#### The SMOTE technique definitely improves most of the model's performance especially the XGBoost and Random Forest classifier which are the best models for this classification task based on the models tested
